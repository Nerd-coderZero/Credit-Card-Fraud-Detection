```python
# credit_card_fraud_detection.ipynb

# Import necessary libraries
import pandas as pd
import joblib
import sys 
import os

# Add the src directory to the Python path
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

from preprocessing import load_data, preprocess_data
from modeling import build_model
from evaluation import evaluate_model

# Load and preprocess the data
print("Loading data...")
data = load_data('../data/creditcard.csv')
print("Data loaded. Preprocessing data...")
X_train_smote, X_test, y_train_smote, y_test = preprocess_data(data)

# Check data shapes and distributions
print(f"Shape of X_train_smote: {X_train_smote.shape}")
print(f"Shape of y_train_smote: {y_train_smote.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_test: {y_test.shape}")

print(f"Distribution in y_train_smote:\n{y_train_smote.value_counts()}")
print(f"Distribution in y_test:\n{y_test.value_counts()}")

# Build and train the best random forest model using hyperparameter tuning
print("Training the best random forest model with hyperparameter tuning...")
best_random_forest_model = build_model('best_random_forest', X_train_smote, y_train_smote)

# Save the best trained model
best_model_file = "../models/best_random_forest_model.pkl"
joblib.dump(best_random_forest_model, best_model_file)
print(f"Best model saved to {best_model_file}")

# Evaluate the best model
print("Evaluating the best Random Forest model...")
evaluate_model(best_random_forest_model, X_test, y_test)

print("Pipeline complete.")
```
